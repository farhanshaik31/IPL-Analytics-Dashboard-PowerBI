Import Libraries 



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

Load Dataset

```python
matches = pd.read_csv("matches.csv")

matches.head()
```

Dataset Overview

```python
print("Rows and Columns:")
print(matches.shape)

print("\nColumns:")
print(matches.columns)

```

Missing Values

```python
matches.isnull().sum()
```

Duplicate Records

```python
matches.duplicated().sum()
```

Remove Duplicates

```python
matches.drop_duplicates(inplace=True)

print("Duplicates Removed")
```

## Exploratory Data Analysis ( EDA )

Winning Teams

```python
top_teams = matches['winner'].value_counts()

print(top_teams)
```

Top Winning Teams Chart


```python
plt.figure(figsize=(10,5))

top_teams.plot(kind='bar')

plt.title("Top Winning IPL Teams")
plt.xlabel("Teams")
plt.ylabel("Wins")

plt.show()
```

Matches Per Season

```python
season_matches = matches['season'].value_counts()

print(season_matches)
```

Season Trend Chart

```python
plt.figure(figsize=(10,5))

season_matches.plot(kind='line', marker='*')

plt.title("Matches Per Season")
plt.xlabel("Season")
plt.ylabel("Matches")

plt.show()
```

Venue Analysis

```python
venue_analysis = matches['venue'].value_counts().head(10)

print(venue_analysis)
```

Venue Chart

```python
plt.figure(figsize=(12,5))

venue_analysis.plot(kind='bar')

plt.title("Top IPL Venues")

plt.show()
```

Toss Winner Analysis

```python
matches['toss_winner'].value_counts().head(10)
```

Toss Winner Chart

```python
plt.figure(figsize=(8,8))

matches['toss_winner'].value_counts().head(10).plot(
    kind='pie'
)

plt.ylabel("")
plt.title("Toss Winner Distribution")

plt.show()
```

# Statistical Analysis

Team Wins Statistics

```python
wins = matches['winner'].value_counts()

print("Mean:", wins.mean())
print("Median:", wins.median())
print("Variance:", wins.var())
print("Standard Deviation:", wins.std())
```
